```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/full_feature_data.csv')
```

```python
df.head()
```

```python
#scatter plot of posts vs. likes and likes vs. comments as subplots

#posts vs. likes
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].scatter(df['Posts'], df['likes'], color='blue')
ax[0].set_title('Posts vs. Likes')
ax[0].set_xlabel('Posts')
ax[0].set_ylabel('Likes')

#likes vs. comments
ax[1].scatter(df['likes'], df['comments'], color='red')
ax[1].set_title('Likes vs. Comments')
ax[1].set_xlabel('Likes')
ax[1].set_ylabel('Comments')

plt.tight_layout()
plt.show()
```

```python
#scatter plot of Posts vs. followers
plt.figure(figsize=(8, 6))
plt.scatter(df['Posts'], df['followers'], color='green')
plt.title('Posts vs. Followers')
plt.xlabel('Posts')
plt.ylabel('Followers')
plt.show()
```

```python
#Avg_Likes_Per_Post vs. Posts
plt.figure(figsize=(8, 6))
plt.scatter(df['Avg_Likes_Per_Post'], df['Posts'], color='purple')
plt.title('Avg Likes Per Post vs. Posts')
plt.xlabel('Avg Likes Per Post')
plt.ylabel('Posts')
plt.show()
```

```python
CK = df[df["business_entity_doing_business_as_name"] == "Calvin Klein"]
```

```python
CK.loc[:-12]
```

```python
CK = CK.loc[-:]
plt.plot(pd.to_datetime(CK["period_end_date"]),CK['main_metrics'], color='blue')
plt.title('Calvin Klein')
plt.xlabel('Period End Date')
plt.ylabel('Main Metrics')
#add standard deviation countour lines
plt.fill_between(pd.to_datetime(CK["period_end_date"]), CK['main_metrics']-CK['main_metrics'].std(), CK['main_metrics']+CK['main_metrics'].std(), color='blue', alpha=0.2)
plt.show()
```
